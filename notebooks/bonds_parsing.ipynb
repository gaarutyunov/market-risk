```python
from datetime import date
from pathlib import Path

import aiohttp
import aiomoex
import pandas as pd
from aiomoex import client
from aiomoex.request_helpers import make_url, get_table, make_query
import datatable as dt
from datatable import f

from glob import glob

__BONDS_FOLDER__ = Path('..', 'data', 'bonds')
```

```python
url = make_url(engine="stock", market="bonds", ending="boards")

async with aiohttp.ClientSession() as session:
    iss = client.ISSClient(session, url)

    data = await iss.get()
    boards = get_table(data, "boards")

    boards_df = pd.DataFrame(boards)
```

```python
traded_boards = boards_df[boards_df.is_traded == 1]
traded_boards
```

```python
check = pd.read_pickle(str(__BONDS_FOLDER__.joinpath('ofz_pd.pickle')))
```

```python
check.shape
```

```python
for security in check.SECID.values:
    url = make_url(engine="stock", market="bonds", board="TQOB", security=security, history=True)
    query = make_query(start='2016-01-01', end='2022-01-01')

    async with aiohttp.ClientSession() as session:
        iss = client.ISSClient(session, url, query)

        data = await iss.get_all()
        history = get_table(data, "history")
        if len(history) == 0:
            continue

        bond_df = pd.DataFrame(history)
        bond_df.to_csv(str(__BONDS_FOLDER__.joinpath('history', security + '.csv')), header=True, index=False)
```
